```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df_pre = pd.read_csv('raiffeisen.csv')
df_pre.head()
```

```python
df = df_pre.drop(columns=['office_id'])
df = pd.get_dummies(df)
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df.head()
```

```python
df.drop(columns=[' latitude', ' longitude'], inplace=True)
```

```python
df.head()
```

```python
from sklearn.cluster import DBSCAN
```

```python
clustering = DBSCAN(eps=3, min_samples=2).fit(df)
clustering.labels_
```

```python
df = pd.read_csv('raiffeisen.csv')
df.head()
```

```python
msk = df[df[' city'] == ' г.Москва']
```

```python
msk.drop(columns=['office_id', ' city', ' latitude', ' longitude'], inplace=True)
msk.head()
```

```python
msk = pd.get_dummies(msk)
msk.head()
```

```python
msk_scaler = StandardScaler()
msk = pd.DataFrame(msk_scaler.fit_transform(msk), columns=msk.columns)
```

```python
msk_clustering = DBSCAN(eps=5).fit(msk)
msk_clustering.labels_
```

```python
msk['labels'] = msk_clustering.labels_
```
