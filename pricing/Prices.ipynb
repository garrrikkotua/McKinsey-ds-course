```python
def appl(x, price):
    if x < price:
        return price
    return x
```

```python
from decimal import Decimal
```

```python
def count(cl, price):
    tmp = data[data['customer_id'].isin(cl)].copy()
    arr1 = np.array(tmp[tmp['price_per_unit'] < price]['price_per_unit']) - price
    arr2 = np.array(tmp[tmp['price_per_unit'] < price]['units'])
    loss = np.sum(arr1 * arr2)
    new_prices = np.array(tmp['price_per_unit'].apply(appl, args=(price, )))
    units = np.array(tmp['units'])
    potential_revenue = np.sum(new_prices * units)
    previous_revenue = np.sum(tmp['revenue'])
    num_transactions = tmp.count()[0]
    return (round(loss), round(potential_revenue), round(previous_revenue), num_transactions, round(loss / num_transactions))
```

```python
with open('economy.csv', 'w') as f:
    f.write('Cluster, '+ 'Loss, ' + 'Potential_Revenue, ' + 'Real_Revenue, ' + 'Num_transactions, ' + 'Avg_trans_loss')
    f.write('\n')
    for cl, pr, l in zip(clusters, prices, letters):
        f.write(l + ', ')
        f.write(', '.join([locale.format("%d", val, grouping=True) for val in count(cl, pr)]))
        f.write('\n')
```
