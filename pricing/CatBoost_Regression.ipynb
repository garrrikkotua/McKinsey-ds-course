```python
from catboost import CatBoostRegressor
from catboost import Pool
```

```python
class ProfitObjective(object):
    
    def __init__(self, data):
        self.data = data
        
    def calc_ders_range(self, approxes, targets, weights):
        profit = np.array([])
        approxes = np.asarray(approxes)
        targets = np.asarray(targets)
        mse = np.mean((approxes - targets) ** 2)
        result = np.array([])
        for index in range(len(targets)):
            np.append(result, approxes[index] * self.data['units'][index] - self.data['cost_ifso'][index] - self.data['cost_dir'][index] - 
                         self.data['cost_overhead'][index] - self.data['cost_commercial'][index] - 
                         self.data['custom_duties'][index])
        return result + mse
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
num_cols = data.columns[data.dtypes.apply(lambda c: np.issubdtype(c, np.number))]
data[num_cols] = scaler.fit_transform(data[num_cols])
data.head()
```

```python
train, test, y_train, y_test = train_test_split(data, price)
```

```python
data.dtypes
```

```python
model = CatBoostRegressor(loss_function='RMSE', cat_features=['market', 'contract_type', 'incoterms', 
                                                                            'country', 'region_id', 'CRM_client_type', 
                                                                            'CRM_client_segment', 'CRM_client_industry'], 
                         iterations=30, max_depth=7)
```

```python
 model.fit(train, y_train, eval_set=(test, y_test))
```

```python
model.feature_importances_
```

```python
dic = {a:b for a, b in zip(model.feature_names_, model.feature_importances_ / model.feature_importances_.max())}
```
