```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df_pre = pd.read_csv('raiffeisen.csv')
df_pre.head()
```

```python
df = df_pre.drop(columns=['office_id'])
df = pd.get_dummies(df)
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df.head()
```

```python
from sklearn.cluster import DBSCAN
```

```python
clustering = DBSCAN(eps=2, min_samples=2).fit(df)
clustering.labels_
```

```python
df = pd.DataFrame(scaler.inverse_transform(df), columns=df.columns)
```

```python
df['labels'] = clustering.labels_
```

```python
df[df['labels'] == 0]
```

```python
df_pre['labels'] = clustering.labels_
```

```python
df_pre[df['labels'] == 8]
```

```python
df_pre[df_pre[' city'] == ' г.Казань']
```

```python
distortions = []
for i in range(1, 11):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km.fit(df)
    distortions.append(km.inertia_)

# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

```python
km = KMeans(
        n_clusters=3, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
km.fit(df)
```

```python
peter = df_pre[df_pre[' city'] == ' г.Санкт-Петербург']
```

```python
peter
```

```python
peter_sc = pd.get_dummies(peter)
```

```python
peter_sc.drop(columns=[' city_ г.Санкт-Петербург'], inplace=True)
```

```python
scaler_pt = StandardScaler()
peter_sc = pd.DataFrame(scaler_pt.fit_transform(peter_sc), columns=peter_sc.columns)
```

```python
distortions = []
for i in range(1, 11):
    km_pt = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )
    km_pt.fit(peter_sc)
    distortions.append(km_pt.inertia_)

# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```
